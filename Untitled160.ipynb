```python
import os
import pandas as pd
import numpy as np
import json, yaml, pickle, matplotlib.pyplot as plt, seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from tensorflow.keras.models import load_model

# ------------------------------------------------------------
# 1ï¸â£ PATH CONFIG
# ------------------------------------------------------------
DATA_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Solar Panel Efficiency Optimizer\archive"
GEN_FILE = os.path.join(DATA_DIR, "Plant_2_Generation_Data.csv")
WEATHER_FILE = os.path.join(DATA_DIR, "Plant_2_Weather_Sensor_Data.csv")

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Solar Panel Efficiency Optimizer\results"
MODEL_FILE = os.path.join(OUTPUT_DIR, "hybrid_SolarSense_model.h5")
SCALER_Y = os.path.join(OUTPUT_DIR, "scaler_y.pkl")
CONFIG_FILE = os.path.join(OUTPUT_DIR, "hybrid_SolarSense_config.yaml")

os.makedirs(OUTPUT_DIR, exist_ok=True)

# ------------------------------------------------------------
# 2ï¸â£ LOAD MODEL, SCALERS & CONFIG
# ------------------------------------------------------------
print("[INFO] Loading model and config...")
model = load_model(MODEL_FILE)
scaler_y = pickle.load(open(SCALER_Y, "rb"))

with open(CONFIG_FILE, "r") as f:
    config = yaml.safe_load(f)
selected_features = config.get("selected_features", [])
print("[INFO] Model expects features:", selected_features)

# ------------------------------------------------------------
# 3ï¸â£ LOAD & MERGE DATA
# ------------------------------------------------------------
print("[INFO] Loading Plant 2 generation + weather data...")
gen = pd.read_csv(GEN_FILE, parse_dates=["DATE_TIME"], dayfirst=False)
weather = pd.read_csv(WEATHER_FILE, parse_dates=["DATE_TIME"], dayfirst=False)
df = pd.merge(gen, weather, on="DATE_TIME", how="inner")
print("[INFO] Merged shape:", df.shape)

# ------------------------------------------------------------
# 4ï¸â£ FEATURE ENGINEERING
# ------------------------------------------------------------
df["EFFICIENCY"] = np.where(df["DC_POWER"] > 0, df["AC_POWER"] / df["DC_POWER"], 0)
df["HOUR"] = df["DATE_TIME"].dt.hour
df["DAY"] = df["DATE_TIME"].dt.day
df["MONTH"] = df["DATE_TIME"].dt.month

# Select only the relevant columns
active_features = [f for f in selected_features if f in df.columns]
print("[INFO] Using features for prediction:", active_features)

df = df.dropna(subset=active_features + ["EFFICIENCY"])
X = df[active_features].values
y = df["EFFICIENCY"].values.reshape(-1, 1)

# ------------------------------------------------------------
# 5ï¸â£ LOCAL SCALING FOR ACTIVE FEATURES
# ------------------------------------------------------------
local_scaler_x = MinMaxScaler()
X_scaled = local_scaler_x.fit_transform(X)
print(f"[INFO] Scaled input shape: {X_scaled.shape}")

# ------------------------------------------------------------
# 6ï¸â£ PREDICTION
# ------------------------------------------------------------
print("[INFO] Running prediction...")
y_pred_scaled = model.predict(X_scaled)
y_pred = scaler_y.inverse_transform(y_pred_scaled)

rmse = np.sqrt(mean_squared_error(y, y_pred))
r2 = r2_score(y, y_pred)
print(f"[RESULT] Hybrid AIS+DA | RMSE={rmse:.4f} | RÂ²={r2:.4f}")

# ------------------------------------------------------------
# 7ï¸â£ SAVE RESULTS
# ------------------------------------------------------------
df["Predicted_Efficiency"] = y_pred
out_csv = os.path.join(OUTPUT_DIR, "hybrid_SolarSense_P2_predictions.csv")
df.to_csv(out_csv, index=False)

summary = {
    "Dataset": "Plant_2 (Merged)",
    "Selected Features": active_features,
    "Records": len(df),
    "RMSE": float(rmse),
    "R2": float(r2),
    "Output File": out_csv
}
with open(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_P2_results.json"), "w") as f:
    json.dump(summary, f, indent=4)

# ------------------------------------------------------------
# 8ï¸â£ VISUALIZATIONS
# ------------------------------------------------------------
plt.figure(figsize=(8,5))
plt.plot(y[:200], label="Actual", color="green")
plt.plot(y_pred[:200], label="Predicted", color="orange", linestyle="--")
plt.title("Hybrid AIS+DA: Plant 2 Actual vs Predicted Efficiency")
plt.xlabel("Samples"); plt.ylabel("Efficiency")
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_P2_comparison_graph.png"), dpi=300)
plt.show()

plt.figure(figsize=(8,5))
plt.plot(y_pred, color="royalblue", linewidth=1.5)
plt.title("Predicted Efficiency Trend (Plant 2)")
plt.xlabel("Samples"); plt.ylabel("Predicted Efficiency")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_P2_prediction_graph.png"), dpi=300)
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x=["RMSE", "R2"], y=[rmse, r2], palette=["crimson", "seagreen"])
plt.title("Hybrid AIS+DA (Plant 2) Performance Metrics")
for i, v in enumerate([rmse, r2]):
    plt.text(i, v + 0.01, f"{v:.3f}", ha="center", fontsize=10, fontweight="bold")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_P2_result_graph.png"), dpi=300)
plt.show()

# ------------------------------------------------------------
# 9ï¸â£ SUMMARY
# ------------------------------------------------------------
print("\nâ Prediction complete! All files saved in:")
print(OUTPUT_DIR)
for f in os.listdir(OUTPUT_DIR):
    if "P2" in f:
        print(" -", f)
```
