```python
# ============================================================
# 🌞 SolarSense | Hybrid AIS + DA Optimized Solar Panel Efficiency Predictor (Final)
# ============================================================

import os
import pandas as pd
import numpy as np
import json, yaml, pickle, random
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import warnings
warnings.filterwarnings("ignore")

# ------------------------------------------------------------
# 1️⃣ CONFIGURATION
# ------------------------------------------------------------
DATA_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Solar Panel Efficiency Optimizer\archive"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Solar Panel Efficiency Optimizer\results"
os.makedirs(OUTPUT_DIR, exist_ok=True)

files = {
    "p1_gen": os.path.join(DATA_DIR, "Plant_1_Generation_Data.csv"),
    "p1_weather": os.path.join(DATA_DIR, "Plant_1_Weather_Sensor_Data.csv"),
    "p2_gen": os.path.join(DATA_DIR, "Plant_2_Generation_Data.csv"),
    "p2_weather": os.path.join(DATA_DIR, "Plant_2_Weather_Sensor_Data.csv"),
}

# ------------------------------------------------------------
# 2️⃣ LOAD & MERGE
# ------------------------------------------------------------
def load_merge(gen_file, weather_file):
    gen = pd.read_csv(gen_file, parse_dates=['DATE_TIME'], dayfirst=False)
    weather = pd.read_csv(weather_file, parse_dates=['DATE_TIME'], dayfirst=False)
    df = pd.merge(gen, weather, on="DATE_TIME", how="inner")
    if 'PLANT_ID_x' in df.columns:
        df['PLANT_ID'] = df['PLANT_ID_x']
    return df

df = pd.concat([
    load_merge(files["p1_gen"], files["p1_weather"]),
    load_merge(files["p2_gen"], files["p2_weather"])
], ignore_index=True)
print("[INFO] Dataset Loaded:", df.shape)

# ------------------------------------------------------------
# 3️⃣ FEATURE ENGINEERING
# ------------------------------------------------------------
df["EFFICIENCY"] = np.where(df["DC_POWER"] > 0, df["AC_POWER"] / df["DC_POWER"], 0)
df["HOUR"] = df["DATE_TIME"].dt.hour
df["DAY"] = df["DATE_TIME"].dt.day
df["MONTH"] = df["DATE_TIME"].dt.month

features = ["AMBIENT_TEMPERATURE", "MODULE_TEMPERATURE", "IRRADIATION", "HOUR", "DAY", "MONTH"]
target = "EFFICIENCY"

df = df.dropna(subset=features + [target])
X, y = df[features].values, df[target].values.reshape(-1, 1)

sx, sy = MinMaxScaler(), MinMaxScaler()
X_s, y_s = sx.fit_transform(X), sy.fit_transform(y)
X_train, X_test, y_train, y_test = train_test_split(X_s, y_s, test_size=0.2, shuffle=False)

# ------------------------------------------------------------
# 4️⃣ HYBRID AIS + DA OPTIMIZATION
# ------------------------------------------------------------
def AIS_feature_selection(X, y, n_clones=5):
    n_features = X.shape[1]
    weights = np.random.rand(n_features)
    for _ in range(n_clones):
        mutation = np.random.normal(0, 0.1, n_features)
        candidate = np.clip(weights + mutation, 0, 1)
        if np.mean(candidate) > np.mean(weights):
            weights = candidate
    mask = weights > np.median(weights)
    return mask.astype(int)

mask = AIS_feature_selection(X_train, y_train)
X_train_sel, X_test_sel = X_train[:, mask == 1], X_test[:, mask == 1]
selected_features = np.array(features)[mask == 1]
print("[AIS] Selected Features:", selected_features)

def build_model(input_dim, params):
    model = Sequential([
        Dense(params["hidden1"], activation='relu', input_shape=(input_dim,)),
        Dropout(params["dropout"]),
        Dense(params["hidden2"], activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=params["lr"]), loss='mse')
    return model

def evaluate_model(X_train, X_test, y_train, y_test, params):
    model = build_model(X_train.shape[1], params)
    model.fit(X_train, y_train, epochs=10, batch_size=64, verbose=0)
    preds = model.predict(X_test, verbose=0)
    rmse = np.sqrt(mean_squared_error(y_test, preds))
    return rmse, model

def DA_optimize(X_train, X_test, y_train, y_test, iterations=6):
    best_rmse = np.inf
    best_params, best_model = None, None
    for i in range(iterations):
        params = {
            "hidden1": random.randint(32, 128),
            "hidden2": random.randint(16, 64),
            "lr": random.uniform(0.0005, 0.01),
            "dropout": random.uniform(0.1, 0.3)
        }
        rmse, model = evaluate_model(X_train, X_test, y_train, y_test, params)
        print(f"[DA] Iter {i+1:02d} | RMSE={rmse:.4f} | Params={params}")
        if rmse < best_rmse:
            best_rmse, best_params, best_model = rmse, params, model
    return best_model, best_params, best_rmse

model, params, best_rmse = DA_optimize(X_train_sel, X_test_sel, y_train, y_test)

# ------------------------------------------------------------
# 5️⃣ FINAL TRAINING (with history captured)
# ------------------------------------------------------------
es = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
history = model.fit(
    X_train_sel,
    y_train,
    epochs=30,
    batch_size=64,
    verbose=1,
    validation_split=0.2,
    callbacks=[es]
)

y_pred = model.predict(X_test_sel)
y_pred_inv, y_test_inv = sy.inverse_transform(y_pred), sy.inverse_transform(y_test)
r2 = r2_score(y_test_inv, y_pred_inv)
print(f"[RESULT] Hybrid AIS+DA => RMSE={best_rmse:.4f}, R2={r2:.4f}")

# ------------------------------------------------------------
# 6️⃣ SAVE ARTIFACTS
# ------------------------------------------------------------
model.save(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_model.h5"))
with open(os.path.join(OUTPUT_DIR, "scaler_x.pkl"), "wb") as f: pickle.dump(sx, f)
with open(os.path.join(OUTPUT_DIR, "scaler_y.pkl"), "wb") as f: pickle.dump(sy, f)

config = {
    "selected_features": selected_features.tolist(),
    "params": params,
    "RMSE": float(best_rmse),
    "R2": float(r2)
}
with open(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_config.yaml"), "w") as f: yaml.dump(config, f)
with open(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_results.json"), "w") as f: json.dump(config, f, indent=4)

# ------------------------------------------------------------
# 7️⃣ VISUALIZATIONS
# ------------------------------------------------------------

# Accuracy curve
plt.figure(figsize=(7,5))
if 'loss' in history.history:
    plt.plot(history.history['loss'], label="Train Loss", linewidth=2)
if 'val_loss' in history.history:
    plt.plot(history.history['val_loss'], label="Validation Loss", linewidth=2, linestyle='--')
plt.title("Hybrid AIS+DA Training Curve")
plt.xlabel("Epochs"); plt.ylabel("MSE Loss")
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_accuracy_graph.png"), dpi=300)
plt.show()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[features + [target]].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap (Hybrid AIS+DA)")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_heatmap.png"), dpi=300)
plt.show()

# Comparison Graph
plt.figure(figsize=(8,5))
plt.plot(y_test_inv[:200], label="Actual", color='green')
plt.plot(y_pred_inv[:200], label="Predicted", color='orange', linestyle='--')
plt.title("Hybrid AIS+DA: Actual vs Predicted Efficiency")
plt.xlabel("Samples"); plt.ylabel("Efficiency")
plt.legend(); plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_comparison_graph.png"), dpi=300)
plt.show()

# Prediction Graph
future_steps = 100
future_input = X_test_sel[-future_steps:]
future_pred = model.predict(future_input)
future_pred_inv = sy.inverse_transform(future_pred)
plt.figure(figsize=(8,5))
plt.plot(range(future_steps), future_pred_inv, color='royalblue', linewidth=2)
plt.title("Hybrid AIS+DA Predicted Efficiency (Future Window)")
plt.xlabel("Time Steps"); plt.ylabel("Predicted Efficiency")
plt.grid(True)
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_prediction_graph.png"), dpi=300)
plt.show()

# Result Graph
plt.figure(figsize=(6,4))
sns.barplot(x=["RMSE", "R2"], y=[best_rmse, r2], palette=["crimson", "seagreen"])
plt.title("Hybrid AIS+DA Performance Metrics")
for i, v in enumerate([best_rmse, r2]):
    plt.text(i, v + 0.01, f"{v:.3f}", ha='center', fontsize=10, fontweight='bold')
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, "hybrid_SolarSense_result_graph.png"), dpi=300)
plt.show()

# ------------------------------------------------------------
# 8️⃣ SUMMARY
# ------------------------------------------------------------
print("\n✅ All hybrid model artifacts and graphs saved in:")
print(OUTPUT_DIR)
for f in os.listdir(OUTPUT_DIR):
    print(" -", f)
```
